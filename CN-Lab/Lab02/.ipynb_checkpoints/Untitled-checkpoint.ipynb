```c++14
#include <iostream>
#include <inttypes.h>
#include <bitset>
```

```c++14
typedef uint32_t crc;
#define WIDTH  (8 * sizeof(crc))
```

```c++14
using namespace std;
uint8_t m[] = {0b10101, 0b1100};
crc r = 0;
cout << sizeof(crc) << " : " << WIDTH << endl;
r ^= m[0] << (WIDTH-8);
cout << bitset<32>(m[0] << (WIDTH-8)) << endl;
cout << bitset<8>(m[0]) << endl;
cout << bitset<32>(r) << endl;
```
